```python
#################################
'''
Get data from web 
Updated in 2018-08-11

Please Note: 
"pandas.io.data" is deprecated, replaced by "pandas-datareader" package.

Reference Link: 

"http://pandas-datareader.readthedocs.io/en/latest/remote_data.html"

'''
####################################

from pandas import DataFrame, Series, Panel 
import pandas_datareader.data as web
from datetime import datetime

all_data = {}

start = datetime(2015, 1, 1)
end = datetime(2018, 8, 10)


for ticker in ['AAPL', 'IBM', 'MSFT', 'GOOG']:
    all_data[ticker] = web.DataReader(ticker,'iex', start, end)
    
price = DataFrame({tic:data['close'] 
                   for tic, data in all_data.items()})

volume = DataFrame({tic:data['volume'] 
                    for tic, data in all_data.items()})

```

```python
# See what data I have got
print(all_data)
print(price)
print(volume)
```

```python
returns = price.pct_change()
returns.tail()
```

```python
# calculate correlation between 2 columns of data (company)
print(returns.MSFT.corr(returns.IBM)) 

print(returns.AAPL.corr(returns.GOOG))

print(returns.GOOG.corr(returns.MSFT))

print(returns.AAPL.corr(returns.MSFT))
```

```python
# Panel Data structure in Pandas

from pandas import Panel 
import pandas_datareader.data as web
from datetime import datetime



start = datetime(2016, 1, 1)
end = datetime(2018, 8, 10)


pdata = Panel(dict((stk, web.DataReader(stk,'iex', start, end))
                  for stk in ['AAPL', 'GOOG', 'MSFT', 'DVMT']))

print(pdata)
```
